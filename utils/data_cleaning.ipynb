```python
import json
import re

def clean_text(text):
    """Remove non-ASCII characters and normalize spaces"""
    # Remove non-ASCII characters
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    
    # Replace multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text)
    
    # Strip leading and trailing spaces
    text = text.strip()
    
    return text

def preprocess_data(file_path):
    """Load and preprocess the data"""
    # Load the JSON file
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    # Directly preprocess the data in place
    for item in data:
        # Replace the original text with cleaned versions
        item['ocr_text'] = clean_text(item['ocr_text'])
        item['figurative_reasoning'] = clean_text(item['figurative_reasoning'])
        
        # Clean each category in the list if present
        if 'meme_depressive_categories' in item and isinstance(item['meme_depressive_categories'], list):
            item['meme_depressive_categories'] = [clean_text(category) for category in item['meme_depressive_categories']]
    
    # Generate save path by adding '_preprocessed' before the file extension
    save_path = file_path.replace('.json', '_preprocessed.json')
    
    # Save preprocessed data to JSON
    with open(save_path, 'w', encoding='utf-8') as f:
        json.dump(data, f, indent=2)
    
    return data

# Execute preprocessing on your file
if __name__ == "__main__":
    file_path = "/Users/dhruv/Work/IITD/winter_semester/NLP/Project/mental-health-meme-classification/dataset/Depressive_Data/final/depressive_val_combined.json"  # Replace with your actual depression data file path
    processed_data = preprocess_data(file_path)
    
    # Print confirmation message
    print(f"Preprocessing complete. Processed {len(processed_data)} records.")
    print(f"Output saved to '{file_path.replace('.json', '_preprocessed.json')}'")
```
